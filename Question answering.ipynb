```python
# Simple Question Answering based Text Classification
class SimpleQASystem:
    def __init__(self):
        # Knowledge base
        self.knowledge_base = {
            "what is a large language model?": """A large language model is an AI model trained on
            massive text data to understand and generate human-like language.""",
            "what is gpt?": """GPT is a large language model developed by OpenAI for text generation
            and natural language understanding."""
        }

    def preprocess(self, question):
        # Convert to lowercase for matching
        return question.strip().lower()

    def get_answer(self, question):
        processed_question = self.preprocess(question)
        if processed_question in self.knowledge_base:
            return self.knowledge_base[processed_question]
        else:
            return "Answer not found"

# Create system
qa_system = SimpleQASystem()

# Given Questions
questions = [
    "What is a Large Language Model?",
    "What is GPT?"
]

# Get Answers
for q in questions:
    answer = qa_system.get_answer(q)
    print("Question:", q)
    print("Answer:", answer)
    print("-" * 40)
```
