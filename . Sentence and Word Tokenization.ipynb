```python
# Import the NLTK library (Natural Language Toolkit)
import nltk

# Import the sentence tokenizer function from NLTK
from nltk.tokenize import sent_tokenize

# Download the 'punkt' tokenizer model
nltk.download('punkt')
nltk.download('punkt_tab')

# Input text that we want to split into sentences (LLM topic)
text = ("""LLMs like GPT and BERT can summarize text, answer questions, and translate languages.
        They are widely used in chatbots, virtual assistants, and AI research.
        These models rely on transformers and large datasets to understand context.""")
# Tokenize the text into individual sentences
sentences = sent_tokenize(text)
# Print the list of extracted sentences
print(sentences)
```
