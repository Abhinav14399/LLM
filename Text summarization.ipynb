```python
# Install required libraries
!pip install transformers datasets --quiet
from transformers import BartTokenizer, BartForConditionalGeneration
# Load BART tokenizer and model
tokenizer = BartTokenizer.from_pretrained("facebook/bart-base")
model = BartForConditionalGeneration.from_pretrained("facebook/bart-base")
# LLM-related input text
text = """
Large Language Models (LLMs) are advanced AI systems trained on massive amounts of text data.
Models such as GPT, BERT, and LLaMA are widely used in applications like chatbots,
text summarization, question answering, and content generation.
LLMs are based on transformer architectures and require powerful hardware such as GPUs
to train and fine-tune efficiently.
"""
print("Original Text:\n", text, "\n")
# Tokenize input
inputs = tokenizer([text], max_length=512, return_tensors="pt", truncation=True)
# Generate summary
summary_ids = model.generate(
inputs["input_ids"],
max_length=50,
min_length=20,
length_penalty=2.0,
num_beams=4,
early_stopping=True
)
# Decode summary
summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
print("Generated Summary:\n", summary)
```
